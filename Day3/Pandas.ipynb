```python
test_list = [100,200,300]
pd.Series(data=test_list)
#This is a Series
```

```python
# Not only a list data type but we can use dictionaries also.
dictionary = {"a":100,"b":200,"c":300}
pd.Series(data=dictionary)
```

### Slicing

```python
# We can slice the dataframe as we did with strings, but here things go a little differently.
# Slicing here is done with rows in Dataframe.
df.iloc[100:111]
```

```python
# Like iloc method there is another method for slicing dataframes i.e. is loc
import numpy as np
s = pd.Series(np.nan, index=[49,48,47,46,45, 1, 2, 3, 4, 5])
s.head(5)
```
