```
#PIE CHART
import matplotlib.pyplot as plt
labels=['a','b','c','d']
size=[20,30,40,10]
color=['r','g','b','y']
e=[0,0,0,0.2]
plt.pie(size,labels=labels,colors=color,explode=e,shadow=True,autopct='%0.1f%%')
plt.legend(['a','b','c','d'],loc='upper right',bbox_to_anchor=(1,0,0.5,1))
plt.show()
```

```
#BAR GRAPH
import matplotlib.pyplot as plt
plt.bar(1,5)
plt.bar(2,10)
plt.bar(3,9)
plt.show()
```

```
#Linear Graph
import matplotlib.pyplot as plt
fig,a = plt.subplots(2,2)
plt.figure(1)
a[0,0].set(xlabel="Overs",ylabel="runs")
a[0,1].set(xlabel="Overs",ylabel="runs")
a[1,1].set(xlabel="Overs",ylabel="runs")
a[1,0].set(xlabel="Overs",ylabel="runs")
a[0,0].plot([1,4,8],[2,4,6],'purple')
a[0,0].plot([1,4,9],[1,4,7],'b')
a[0,0].legend(['KKR','MI'],loc='best')
a[0,1].plot([1,4,8],[2,4,6],'y')
a[0,1].plot([1,4,9],[1,4,7],'b')
a[0,1].legend(['CSK','MI'],loc='best')
a[1,0].plot([1,4,8],[2,4,6],'y')
a[1,0].plot([1,4,9],[1,4,9],'r')
a[1,0].legend(['CSK','RR'],loc='best')
a[1,1].plot([1,4,8],[2,4,6],'y')
a[1,1].plot([1,4,9],[1,4,7],'b')
a[1,1].legend(['CSK','MI'],loc='best')
plt.show()
```
